```python
import matplotlib.pyplot as plt

# Set up figure
fig, axes = plt.subplots(1, 3, figsize=(18,6))
fig.suptitle('Toronto Collision Severity Dashboard: User, Environment, and Time Factors', fontsize=16, fontweight='bold')

# -------------------------------
# USER FACTORS – Involvement Type
# -------------------------------
involvement = ["Cyclist","Motorcycle Driver","Moped Driver","Pedestrian","Passenger","Truck Driver"]
odds_ratio = [34.4, 23.4, 20.8, 13.8, 0.66, 0.54]

axes[0].barh(involvement, odds_ratio, color=['tomato']*4 + ['lightgray']*2)
axes[0].axvline(x=1, color='gray', linestyle='--')
axes[0].set_xlabel('Odds Ratio (vs Drivers)')
axes[0].set_title('User Factors\n(Severity Risk by Involvement Type)', fontsize=13)
axes[0].invert_yaxis()
axes[0].text(10, -0.5, "Unprotected users\nface 13–34× higher odds", color='black', fontsize=10)

# -------------------------------
# ENVIRONMENTAL FACTORS – Lighting Condition
# -------------------------------
lighting = ["Daylight","Dusk (Artificial)","Dark (Artificial)","Dark (Natural)"]
odds_ratio_light = [1.0, 1.9, 0.94, 0.89]

axes[1].bar(lighting, odds_ratio_light, color=['lightgray','gold','gray','dimgray'])
axes[1].axhline(y=1, color='gray', linestyle='--')
axes[1].set_ylabel('Odds Ratio (vs Daylight)')
axes[1].set_title('Environmental Factors\n(Impact of Lighting Condition)', fontsize=13)
axes[1].text(0.5, 1.95, "Dusk lighting nearly\ndoubles severity risk", color='black', fontsize=10)

# -------------------------------
# TIME FACTORS – Predicted Probability by Time of Day
# -------------------------------
time_range = ["3–5:59","6–8:59","9–11:59","12–14:59","15–17:59","18–20:59","21–23:59"]
pred_prob_time = [0.69, 0.78, 0.76, 0.74, 0.75, 0.75, 0.72]

axes[2].plot(time_range, [p*100 for p in pred_prob_time],
             marker='o', color='teal', linewidth=2)
axes[2].fill_between(time_range, [p*100 for p in pred_prob_time], color='teal', alpha=0.1)
axes[2].set_ylim(65, 85)
axes[2].set_ylabel('Predicted Probability of Severe Injury (%)')
axes[2].set_title('Time Factors\n(Severity Probability by Time of Day)', fontsize=13)
axes[2].text(1.2, 83, "Morning rush hours (6–9 AM)\nhave highest severity", color='black', fontsize=10)
axes[2].grid(alpha=0.3)

# -------------------------------
# Adjust spacing and export
# -------------------------------
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.savefig('Toronto_Collision_Severity_Dashboard.png', dpi=300, bbox_inches='tight')
plt.show()
```
